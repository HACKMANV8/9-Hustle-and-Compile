```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import os


model_name = "cardiffnlp/twitter-roberta-base-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
sentiment_classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

def analyze_sentiment(text):
    """Analyze sentiment of the input text"""
    results = sentiment_classifier(text[:512])  
    label_map = {"LABEL_0": "Negative", "LABEL_1": "Neutral", "LABEL_2": "Positive"}
    sentiment_label = label_map.get(results[0]['label'], "Neutral")
    sentiment_score = results[0]['score']
    
    if sentiment_label == "Negative":
        sentiment_score = -sentiment_score
    elif sentiment_label == "Neutral":
        sentiment_score = 0
    
    return sentiment_score, sentiment_label

def analyze_stress(text):
    """Detect stress level based on keywords and linguistic patterns"""
    stress_keywords = ['stress', 'anxious', 'anxiousness', 'burnout', 'overwhelmed', 'tired', 'exhausted', 
                       'depressed', 'depressing', 'frustrated', 'frustrated', 'worried', 'panic', 'fear', 'scared']
    stress_count = sum(1 for keyword in stress_keywords if keyword in text.lower())
    
    if stress_count >= 2:
        return "High"
    elif stress_count == 1:
        return "Medium"
    else:
        return "Low"

def analyze_attention(text):
    """Estimate attention/cognitive load from linguistic complexity"""
    words = text.split()
    sentences = text.count('.') + text.count('!') + text.count('?')
    
    if len(words) == 0:
        return 0.5
    
    avg_word_length = np.mean([len(word) for word in words])
    avg_sentence_length = len(words) / max(sentences, 1)
    
    
    attention_score = (avg_word_length / 10 + avg_sentence_length / 100) / 2
    attention_score = min(max(attention_score, 0), 1)
    
    return attention_score

def analyze_mood(text):
    """Infer overall mood from sentiment and stress"""
    sentiment_score, sentiment_label = analyze_sentiment(text)
    stress_level = analyze_stress(text)
    
    if sentiment_label == "Positive" and stress_level == "Low":
        mood = "Happy / Content"
    elif sentiment_label == "Negative" and stress_level == "High":
        mood = "Anxious / Stressed"
    elif sentiment_label == "Negative":
        mood = "Sad / Discouraged"
    elif stress_level == "High":
        mood = "Overwhelmed"
    elif stress_level == "Medium":
        mood = "Thoughtful / Reflective"
    else:
        mood = "Calm / Neutral"
    
    return mood

def analyze_behavioral_markers(text):
    """Extract psycholinguistic behavioral markers"""
    markers = {}
    
    
    words = text.lower().split()
    markers['word_repetition'] = len(words) - len(set(words)) if len(words) > 0 else 0
    
    
    markers['exclamations'] = text.count('!')
    markers['questions'] = text.count('?')
    
    
    markers['avg_word_length'] = np.mean([len(word) for word in words]) if len(words) > 0 else 0
    
    return markers

print("✓ Sentiment analysis pipeline loaded successfully!")

```

```python
import re

def detect_crisis_risk(text):
    text = text.lower().strip()

    # Normalize fillers/slang
    slang = {
        "yaar": "", "pls": "please", "plz": "please", "idk": "i don't know",
        "fr": "for real", "abt": "about", "nth": "nothing"
    }
    for s,r in slang.items():
        text = text.replace(s, r)

    # HIGH RISK patterns (explicit or implicit self‑harm intent)
    high_patterns = [
        r"want to die",
        r"wish (i|you) were dead",
        r"i( might| may)? die( today| soon)?",
        r"feel like dying",
        r"ending my life",
        r"end it all",
        r"kill myself",
        r"self harm",
        r"suicid(al|e)",
        r"no reason to live",
        r"life is pointless",
        r"i don't want to live",
        r"can't live anymore",
        r"i'm done with life",
        r"i don't deserve to live",
        r"if i disappear",
        r"better off dead",
        r"should not exist",
        r"cut myself",
        r"hurt myself",
        r"die",
        r"suicide",
        r"kill"
    ]

    # Medium risk (severe emotional distress, burnout)
    medium_patterns = [
        r"can't handle life",
        r"breaking down",
        r"mentally (exhausted|drained|broken)",
        r"nobody cares",
        r"i feel hopeless",
        r"i give up",
        r"i'm losing it",
        r"no motivation",
        r"hate my life",
        r"can't cope",
    ]

    # Low risk (sadness/stress cues)
    low_patterns = [
        r"sad",
        r"down",
        r"anxious",
        r"tired",
        r"drained",
        r"stress(ed)?",
        r"crying",
        r"low mood",
        r"not okay",
        r"feeling heavy"
    ]

    # Check high → medium → low
    for p in high_patterns:
        if re.search(p, text):
            return "🚨 HIGH RISK – Immediate support needed"

    for p in medium_patterns:
        if re.search(p, text):
            return "⚠️ MEDIUM RISK – Emotional support needed"

    for p in low_patterns:
        if re.search(p, text):
            return "🌥️ LOW RISK – Monitor mood"

    return "✅ No crisis detected"
```
